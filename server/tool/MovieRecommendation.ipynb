Fields selection

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genres', 'overview']]
```

```python
movies
```

```python
movies.loc[:, 'tags'] = movies['overview'] + movies['genres']
movies
```

```python
movies = movies.drop(columns=['overview', 'genres'])
```

```python
movies
```

Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=len(movies), stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
vector.shape
```

Consine similarity computation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

Top similarity selection

```python
top_10_similarity = {}
for i in range(len(movies)):
    # Sort similarities and ignore the first one (since it's the movie itself)
    distances = sorted(list(enumerate(similarity[i])), reverse=True, key=lambda x: x[1])[1:11]
    # Map the movie ids instead of indices
    valid_distances = [(movies.iloc[idx]['id'], sim) for idx, sim in distances if idx < len(movies)]
    top_10_similarity[movies.iloc[i]['id']] = valid_distances
```
