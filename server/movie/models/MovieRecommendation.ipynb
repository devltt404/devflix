Fields selection

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genres', 'overview']]
```

```python
movies
```

```python
movies.loc[:, 'tags'] = movies['overview'] + movies['genres']
movies
```

```python
movies = movies.drop(columns=['overview', 'genres'])
```

```python
movies
```

Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=len(movies), stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
vector.shape
```

Consine similarity computation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

Recommend function

```python
def recommend(movie_id):
    index = movies[movies['id'] == movie_id].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].id)
```

```python
recommend(893370)
```
